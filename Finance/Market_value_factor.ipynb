```python
#get data
import requests
import pandas as pd
import numpy as np
import io
import datetime
```

```python
#抓取股票價格、市值
def Market_value(indate):  
    r = requests.post('https://www.tpex.org.tw/web/stock/aftertrading/daily_mktval/mkt_result.php?l=zh-tw&d='+str(indate)+'&s=0,asc,0&o=csv')
    r = r.text.replace('個股市值排行   \r\n資料日期:'+ str(indate) +'\r\n','').replace('\r','')
    f = io.StringIO(r)
    df = pd.read_csv(f)
    
    #轉成西元
    data_c = indate.split("/")    
    data_c[0]=str(int(data_c[0])+1911)
    data_c = '-'.join(data_c)
    df.loc[:,'Date'] = data_c
    
    #日期當作index
    #df.set_index('Date' , inplace=True)
    
        
    #重新排序colums
    cols = df.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    df = df[cols]
    df = df.iloc[:-2]
    return df
```

```python
#股價須以民國爬取，取得時間列表並轉換為民國
#輸入西元日期ex:datestart='1991-03-15' dateend = '1991-03-18'
def Republic_date(datestart,dateend):
    #取得時間列表
    datestart=datetime.datetime.strptime(datestart,'%Y-%m-%d')
    dateend=datetime.datetime.strptime(dateend,'%Y-%m-%d')
    date_list_r = []
    date_list_c = []
    date_list_r.append(datestart.strftime('%Y-%m-%d'))
    while datestart<dateend:
        datestart+=datetime.timedelta(days=+1) 
        date_list_r.append(datestart.strftime('%Y-%m-%d'))
    #print(date_list_r) 
    
    #西元轉民國
    for d in date_list_r:
        d=d.split("-")
        d[0] = str(int(d[0])-1911)
        d='/'.join(d)
        date_list_c.append(d)
    #print(date_list_c)
    
    return date_list_c
```

```python
#股價資訊存入sql
from sqlalchemy import create_engine
def  Market_value_to_sql(data):
    
    #Market_value抓取股票價格市值
    df = Market_value(data)    
    engine = create_engine("mysql+pymysql://{}:{}@{}/{}?charset={}".format('root', 'mysql', '127.0.0.1:3306', 'test','utf8'))
    
    #創建連結
    con = engine.connect()
    df.to_sql(name='market_value', con=con, if_exists='append', index=False)
    
    #不加close() 'Too many connections'
    con.close()
```

```python
#爬取股價資訊存入sql
date_list=Republic_date('2007-01-01','2019-06-04')
for date in date_list:
    #存入sql
    Market_value_to_sql(date)
    print("正在輸入",date)    
```
