```python
test_file="【時報記者任珮云台北報導】權王台積電 (2330) 昨(24)日除息8元，揭開台股除權息旺季序幕，除息參考價240.5元。終場收在241元，上漲0.5元，小幅度填息，但外資與自營商由買轉賣，合計賣超1.2萬多張。今(25)日早盤，台積電股價仍呈現整理，最低價來到238元，相較除息參考價已轉為貼息。由於台積電股價與美國費半指數呈現高相關性，投資人可以密切關注兩者強弱，研判後續台積電除息行情走勢。國泰台灣加權反1 ETF (00664R) 基金經理人蘇鼎宇表示，聯準會6月會議雖然暗示未來降息可能性，但仍需視G20川習會中美貿易談判結果及美國經濟數據變化，市場不確定性因素仍在，若聯準會7月即展開降息，需留意市場是否以總體經濟弱化進行解讀，導致偏空回應。蘇鼎宇補充，G20川習會前，包含台股在內，全球股市已大漲反映樂觀期待，但目前台股本益比高於過去10年均值約4%，且時序上進入除息季節，以及6、7月份的電子淡季，基本面能否接棒，仍存變數，現階段除操作強勢股外，亦可運用國內第一檔提供加權指數反向報酬的「國泰台灣加權反1」ETF (00664R) ，控制整體大盤風險。"
```

```python
import jieba
#分詞

#jieba.cut 以及 jieba.cut_for_search 返回 generator
jieba_cut = " ".join(jieba.cut(test_file))
print("jieba_cut:",jieba_cut)

jieba_cut_search = " ".join(jieba.cut_for_search(test_file))
#print("jieba.cut_for_search:",jieba_cut_search)

#jieba.lcut 以及 jieba.lcut_for_search 返回 list
jieba_lcut = jieba.lcut(test_file)
#print("jieba_lcut:",jieba_lcut)
```

```python
#載入詞庫
#可以指定自己自定義的辭典，以添加 jieba詞庫裡沒有的詞，以提高正確率
jieba.load_userdict("dict.txt")
jieba_cut = " ".join(jieba.cut(test_file))
print("jieba_cut:",jieba_cut)
```

```python
# 若斷詞過後，與我們想要的詞有些微出入，可自創建詞庫。
#這邊試著將台積電加入自創建詞庫new_dict.txt
jieba.load_userdict("new_dict.txt")
jieba_cut = " ".join(jieba.cut(test_file))
print("jieba_cut:",jieba_cut)
```

```python
#強制提高詞頻
jieba.suggest_freq('川習會', True)
jieba_cut = " ".join(jieba.cut(test_file))
print("jieba_cut:",jieba_cut)
```

```python
#直接删除詞
jieba.del_word('任珮云')
jieba_cut = " ".join(jieba.cut(test_file))
print("jieba_cut:",jieba_cut)
```

```python
from jieba.analyse import extract_tags
#sentence 文本
#topK 返回幾個TF/IDF權重最大關鍵詞，默認20個
#withWeight 是否返回關鍵詞權重值，默認False
#allowPOS 選擇特定詞性的詞，默認不篩選

jieba.analyse.extract_tags(test_file, topK=20, withWeight=True, allowPOS=())
```

```python
#TextRank
#ref:https://medium.com/@danjtchen/textrank-%E6%96%87%E5%AD%97%E6%8E%A2%E5%8B%98-%E6%89%BE%E5%87%BA%E9%97%9C%E9%8D%B5%E5%AD%97-%E4%BB%A5-%E5%85%AB%E5%8D%A6%E7%89%88%E6%A8%99%E9%A1%8C%E7%82%BA%E4%BE%8B-b16620370872
from jieba.analyse import textrank
jieba.analyse.textrank(test_file, topK=20, withWeight=True, allowPOS=('ns', 'n', 'vn', 'v'))
```
